```python
import pandas as pd
import sqlite3

# Paste GitHub file URL here (must end in .csv)
url = 'https://raw.githubusercontent.com/fahrefai96/UWL---Databases-and-Analytics-Assignment/main/greenfuture_ideas_dataset.csv'

# Load the dataset
data = pd.read_csv(url)

# Creating a temporary in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Loading the DataFrame 'data' into a SQL table named 'ideas'
data.to_sql('ideas', conn, index=False, if_exists='replace')

# Query the ideas table to view all the data, limit it to 5 records and their results
pd.read_sql("SELECT * FROM ideas LIMIT 5", conn)




```

```python
# Query the ideas table to view Employee ID and thier Office Location, limit by 5 records and display results

pd.read_sql('SELECT "Employee ID", "Office Location" FROM ideas LIMIT 5', conn)
```

```python
# Insert a new record into the 'ideas' table

cursor = conn.cursor()
cursor.execute("""
INSERT INTO ideas (`Employee ID`, `Office Location`, `Department`, `Idea Submission Date`, `Idea Title`, `Idea Category`, `Number of Votes`, `Collaboration Status`, `Approval Status`, `Implementation Status`, `Security Concern Flag`)
VALUES ('1234abcd', 'USA', 'Technology', '20/05/2025', 'Idea 105', 'Waste Management', 120, 'Single Contributor', 'Pending', 'Not Started', 'No')
""")
conn.commit()

pd.read_sql('SELECT "Employee ID", "Office Location" FROM ideas WHERE "Employee ID" = "1234abcd"', conn)
```

```python
# Update a record where Employee ID is '587b45d8' to set the Approval Status to Approved
cursor.execute("""
UPDATE ideas
SET `Approval Status` = 'Approved'
WHERE `Employee ID` = '587b45d8'
""")
conn.commit()

pd.read_sql('SELECT * FROM ideas WHERE "Employee ID" = "587b45d8"', conn)
```

```python
# Delete a record where Employee ID is '333fff26'
cursor.execute("""
DELETE FROM ideas WHERE `Employee ID` = '333fff26'
""")
conn.commit()
```

```python
#  Query using Mathematical Expressions on Number of Votes

pd.read_sql("""
SELECT
  `Idea Title`,
  `Number of Votes`,
  `Number of Votes` + 10 AS `Votes_Plus_10`,
  `Number of Votes` * 2 AS `Votes_Double`,
  `Number of Votes` - 5 AS `Votes_Minus_5`
FROM ideas
LIMIT 5
""", conn)
```

```python
#Group the ideas by whether there is a security concern ("Yes"/"No") and counts how many fall into each category

pd.read_sql("""
SELECT
  `Security Concern Flag`,
  COUNT(*) AS `Total`
FROM ideas
GROUP BY `Security Concern Flag`
""", conn)
```

```python
# Apply aggregate function to calculate the average number of votes and show it with Idea Category

pd.read_sql("SELECT `Idea Category`, AVG(`Number of Votes`) AS `Average Votes` FROM ideas GROUP BY `Idea Category`", conn)
```

```python
#Find the earliest and latest submission dates in the dataset using MIN and MAX
pd.read_sql("""
SELECT
  MIN(`Idea Submission Date`) AS `Earliest Submission`,
  MAX(`Idea Submission Date`) AS `Latest Submission`
FROM ideas
""", conn)
```

```python
#Subtract 50 from the Number of Votes column for each row, create a new column called Adjusted Votes and limit to 5 records.

pd.read_sql("""
SELECT
    `Employee ID`,
    `Number of Votes` - 50 AS `Adjusted Votes`
FROM ideas LIMIT 5;
""", conn)
```
